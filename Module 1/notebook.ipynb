```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
taxi_df = pd.read_parquet("../data/green_tripdata_2025-11.parquet")
lookup = pd.read_csv("../data/taxi_zone_lookup.csv")
```

```python
taxi_df.shape, lookup.shape
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')

engine.connect()
```

```python
lookup.to_sql(name='taxi_zone_lookup', con=engine, if_exists='replace', index=False)

taxi_df.to_sql(name='green_tripdata', con=engine, if_exists='replace', index=False, chunksize=10000)
```
